# CLEANING — MATCHES DATASET

```python
# Import Required Libraries
import pandas as pd
import numpy as np
```

```python
# Load dataset
df = pd.read_excel('matches (1).xlsx')
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df.drop_duplicates()
```

```python
# checking null values

print(df.isnull().sum())
```

```python
# 2. Convert datatype
df['date'] = pd.to_datetime(df['date'])
```

```python
# 3. HANDLE umpire3 (IMPORTANT FIX)
df['umpire3'].fillna('Not Available', inplace=True)
```

```python
# 4. Handle missing values

# Categorical columns
cat_cols = df.select_dtypes(include='object').columns

for col in cat_cols:
    if col != 'umpire3':   # skip umpire3
        df[col].fillna(df[col].mode()[0], inplace=True)

# Numerical columns
num_cols = df.select_dtypes(include=['int64','float64']).columns

for col in num_cols:
    df[col].fillna(df[col].median(), inplace=True)
```

```python
# 5. Remove negative values
for col in num_cols:
    df[col] = df[col].apply(lambda x: abs(x))
```

```python
# 6. Handle outliers (IQR)
for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    df[col] = np.where(df[col] < lower, df[col].median(), df[col])
    df[col] = np.where(df[col] > upper, df[col].median(), df[col])
```

```python
# 7. Feature Engineering

# Extract year
df['year'] = df['date'].dt.year

# Toss impact column
df['toss_win_match_win'] = np.where(df['toss_winner'] == df['winner'], 1, 0)
```

```python
# 8. Save cleaned data to Excel
df.to_excel('matches_cleaned.xlsx', index=False)

```
